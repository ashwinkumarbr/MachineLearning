```python
b1['RelDate'].head(5)
```

lets first correct the formats, along with identying any underlying issues.

```python
# this funtion, will take the Movie number for reference, Releasedate as string, and expected format of the date.
# returns the date along with printing any wrong format date , if cannot be valdiated.
import datetime
def val_year(No, RelDate,f):
    try:
        dt = datetime.datetime.strptime(RelDate, f)
        return dt.date()
    except:
        print(f"Date '{RelDate}' for the movie  '{No}'")
        return RelDate
```

```python
# format of RelDate in b1 source is Jul-4-2014
b1_f='%b-%d-%Y'

b1['RelDate'] = b1.apply(lambda x: val_year(x.No,x.RelDate,b1_f), axis=1)
```

```python
# format of RelDate in b2 source is 4-Jul-14
b2_f = '%d-%b-%y'
b2['RelDate'] = b2.apply(lambda x: val_year(x.No,x.RelDate,b2_f), axis=1)
```

    # In  order to map the non numeric release times
we will use this below.
Festive Season
    FS is 1,
Holiday Season
    HS is 2,
Long Weekend
    LW is 3,
Normal
    N is 4

```python
# function takes x as the releasetime and returns a equivalent number
def map_ReleaseTime(x):
    switcher = { 
        "FS": 1,
        "HS": 2,
        "LW": 3,
        "N": 4,
    }
    return switcher.get(x, x) 
```

```python
#invoke the switch function

bdata['ReleaseTime'] = bdata["ReleaseTime"].apply(lambda x: map_ReleaseTime(x))
```

## Q6. Which are the top 10 movies with maximum return of investment (ROI)? 
## Calculate ROI as (Box office collection  Budget) / Budget.

```python
# function to calculate above solution
#Where 
'''
Box = BoxOfficeCollection
Bud = Budget
roi = (Box office collection  Budget) / Budget
'''
def roi(Box,Bud):
    roi=(Box-Bud)/Bud
    return roi
```

```python
#Invoke ROI calculation

bdata['ROI'] = bdata.apply(lambda x: roi(x.BoxOfficeCollection,x.Budget), axis=1 )
```

## Solution to Q6 

```python
# fetch top 10 ROI
bdata.sort_values(['ROI'], ascending=False)[0:10][['No','MovieName','ROI']]
```
